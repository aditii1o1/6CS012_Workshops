Task - 1:

Create a Python program that converts between different units of measurement.

```python
def convert_units():
    """
    Converts between length, weight, and volume units.

    Returns:
        float: converted value
    """
    try:
        print("Select conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Enter choice (1/2/3): ")

        value = float(input("Enter value to convert: "))

        if choice == "1":
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            c = input("Choose option: ")

            if c == "1":
                result = value * 3.28084
                print("Result:", result, "ft")

            elif c == "2":
                result = value / 3.28084
                print("Result:", result, "m")

        elif choice == "2":
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            c = input("Choose option: ")

            if c == "1":
                result = value * 2.20462
                print("Result:", result, "lbs")

            elif c == "2":
                result = value / 2.20462
                print("Result:", result, "kg")

        elif choice == "3":
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            c = input("Choose option: ")

            if c == "1":
                result = value * 0.264172
                print("Result:", result, "gal")

            elif c == "2":
                result = value / 0.264172
                print("Result:", result, "L")

        else:
            print("Invalid option")

    except ValueError:
        print("Invalid input. Please enter numeric value.")

convert_units()
```

Create a Python program that performs various mathematical operations on a list of numbers.

```python
def find_sum(numbers):
    """Returns sum of numbers"""
    return sum(numbers)

def find_average(numbers):
    """Returns average of numbers"""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Returns maximum number"""
    return max(numbers)

def find_min(numbers):
    """Returns minimum number"""
    return min(numbers)

try:
    nums = input("Enter numbers separated by space: ")
    numbers = list(map(float, nums.split()))

    print("Choose operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter choice: ")

    if choice == "1":
        print("Sum =", find_sum(numbers))

    elif choice == "2":
        print("Average =", find_average(numbers))

    elif choice == "3":
        print("Max =", find_max(numbers))

    elif choice == "4":
        print("Min =", find_min(numbers))

    else:
        print("Invalid choice")

except:
    print("Invalid input")
```

#**Nested Lists**

###Exercise 4.3.1 – Flatten Nested List

```python
def flatten(lst):
    result = []
    for sub in lst:
        for item in sub:
            result.append(item)
    return result

print(flatten([[1,2],[3,4],[5]]))
```

###Exercise 4.3.2 – Access Nested Element


```python
def access_nested_element(lst, indices):
    element = lst
    for i in indices:
        element = element[i]
    return element

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))
```

###Exercise 4.3.3 – Sum Nested List

```python
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

###Exercise 4.3.4 – Remove Element from Nested List

```python
def remove_element(lst, elem):
    result = []
    for sub in lst:
        result.append([x for x in sub if x != elem])
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

###Exercise 4.3.5 – Find Maximum in Nested List

```python
def find_max(lst):
    max_val = float('-inf')
    for item in lst:
        if isinstance(item,list):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val,item)
    return max_val

print(find_max([[1,2],[3,[4,5]],6]))
```

###Exercise 4.3.6 – Count Occurrences

```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item,list):
            count += count_occurrences(item,elem)
        else:
            if item == elem:
                count += 1
    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

###Exercise 4.3.7 – Deep Flatten

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item,list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

###Exercise 4.3.8 – Nested List Average

```python
def flatten(lst):
    result=[]
    for item in lst:
        if isinstance(item,list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

def average_nested(lst):
    flat = flatten(lst)
    return sum(flat)/len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```
